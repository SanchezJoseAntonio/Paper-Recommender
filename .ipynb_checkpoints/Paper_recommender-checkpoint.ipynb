```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors
import json
import pandas as pd
from pylatexenc.latex2text import LatexNodes2Text # to turn latex into string
import string
import re
import joblib

```

```python
file_path = './arxiv-metadata-oai-snapshot.json'

# Load the first 5000 papers into a dataframe
data = []
with open(file_path, 'r') as f:
    for i, line in enumerate(f):
        if i >= 5000:  # Change number to adjust sample size
            break
        data.append(json.loads(line))

df = pd.DataFrame(data)
```

In order to process the recommendations, I will check the similarity to the user input to the abstracts. 
The abstracts are a high quality source, so misspellings are assumed to be rare. However, they are highly specialized as well. 
## Pre-processing
The data needs some pre-processing, particularly punctuation and line feeds need to be deleted as they carry no semantic meaning. The same applies to upper case letters, the standalone numbers (i.e 2021, but not COVID-19) and standalone punctuation.

```python
df_cp = df.copy()
```

```python
df_cp["abstract"] =df_cp["abstract"].str.lower()
df_cp["abstract"].head()
```

The abstracts are using Latex " ($(k,\ell)$) " which could cause noise, worsening the recommendations. 

```python
df_cp = df.copy()

converter = LatexNodes2Text()

def clean_abstract(abstract):
    text = converter.latex_to_text(abstract)
    text = text.lower().replace("\n", " ") # Delete line feeds, lowercase everything
    text = " ".join([a.strip(string.punctuation) for a in text.split()]) # Delete punctuation from beginning and end of words
    text = re.sub(r'\b\d+(\.\d+)?\b', '', text) # Delete isolated numbers as they have no meaning
    return text

df_cp['clean_abstract'] = df_cp['abstract'].apply(clean_abstract)

```

# Vectorization and nearest neighbors search

```python
vectorizer = TfidfVectorizer(
    ngram_range=(1,2), # Both one word and two words are extracted as features
    max_df=0.8, # Ignore terms that appear in more than 80% of the documents
    min_df=3, # Ignore terms that appear in less than 3 documents
    max_features=50000, # Limit the vocabulary to the 50000 most important features
    stop_words='english', # Remove common english words that carry no interesting meaning for dimensionality reduction
    sublinear_tf=True, # Lower effect of very common terms by applying log scaling to term frequency
    norm='l2' # Normalize so that length does not cause bias
)
```

This configuration is tailored for scientific abstracts by capturing both individual terms and key phrases, filtering out globally uninformative words, and normalizing for document length so similarity comparisons are not biased by abstract size.

```python
X = vectorizer.fit_transform(df_cp['clean_abstract'])
nn = NearestNeighbors(n_neighbors=10, metric='cosine', algorithm='brute')
nn.fit(X)
```

```python
def recommend_with_scores(query, k=3):
    q_vec = vectorizer.transform([query]) 
    distances, indices = nn.kneighbors(q_vec, n_neighbors=k)
    scores = 1 - distances[0]  # cosine similarity
    return [(idx, float(score)) for idx, score in zip(indices[0], scores)]
```

```python
# Example query
query = "deep learning is based on human learning"
results = recommend_with_scores(query, k=3)

for index, score in results:
    print(f"{df_cp["title"][index]} — score: {score:.4f}") # Show the title instead of the abstract (easier for search)
```

```python
# Save the model for deployment
joblib.dump(X , 'vectorizer.joblib')
```
